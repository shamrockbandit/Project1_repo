```python
#  Notebook exeercising API
freedom_csv_path = "data/indexfreedom.csv"
#happy_csv_path = "data/TBD"
```

```python
# READ freedom.csv

freedom_df = pd.read_csv(freedom_csv_path, encoding="utf-8")
freedom_df
```

```python
# drop rows for 2008-2015 
other_years = freedom_df[ (freedom_df['year'] < 2016) ].index

# Delete these row indexes from dataFrame
only_2016_df= freedom_df.drop(other_years , inplace=True)


freedom_df
```

```python
money_growth_df = freedom_df[['countries','ef_money_growth','hf_score']]
money_growth_df
```

```python
import matplotlib.pyplot as plt
```

```python
# first turn columns into lists
money_growth = money_growth_df["ef_money_growth"].values
hf_score = money_growth_df["hf_score"].values
countries = money_growth_df["countries"].values
```

```python
# create plot
fig, ax = plt.subplots()
ax.scatter(money_growth, hf_score)

plt.title("Money Growth v Human Freedom Score")
plt.xlabel("Money Growth")
plt.ylabel("Human Freedom Score")
```

```python
# independent t-test
import scipy.stats as stats
```
