```python
# droping all NaN's loses the entire data set.
# may need to select which columns we care about data existing 
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html
#freedom_noNan_df = freedom_country_df.dropna()

print(f"Original data shape: {freedom_country_df.shape}")
print(f"No NaN shape: {freedom_country_df.dropna().shape}")
```

```python
# Example select a column we care about having data and drop NaN rows from that 
freedom_clean_df = freedom_country_df.dropna(subset=['ef_score'])
print(f"df shape after selected dropna: {freedom_clean_df.shape} ")

freedom_clean_df.head()
```

```python
#how many years of data are in this set?
# data ranges from 2008 to 2016
freedom_country_df['year'].value_counts()
```

```python
#  Mulit-index groupBy region, year
# Group by Region seems low hanging fruit
# since data ranges from 2008 - 2016 group by year also
# https://www.geeksforgeeks.org/python-pandas-dataframe-groupby/

freeRegionYear_gb = freedom_clean_df.groupby(['region', 'year'])
freeRegionYear_gb.first()
```

```python
# calculate means by region and year groupby
freeRegionYear_mean_df = freeRegionYear_gb.mean()
freeRegionYear_mean_df
```

```python
# can also group by country and year
freeCountryYear_gb = freedom_clean_df.groupby(['countries', 'year'])
freeCountryYear_gb.first()
```

```python
# do summary calcs by country and year
# e.g. get the Means into a df
freeCountryYear_mean_df = freeCountryYear_gb.mean()
freeCountryYear_mean_df
```
