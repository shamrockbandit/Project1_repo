```python
import pandas as pd
import numpy as np
```

```python
#read in the csv
happy_csv = "data/world-happiness-report-2019.csv"
```

```python
#create dataframe
happy_df = pd.read_csv(happy_csv)
happy_df
```

```python
#rename last two columns for clarity 
happy_df.rename(columns={'Log of GDP\nper capita': 'Log of GDP per capita', 'Healthy life\nexpectancy': 'Healthy life expectancy'}, inplace=True)
happy_df
```

```python
#set country as the new index 
happy_df = happy_df.set_index(['Country (region)'])
happy_df
```

```python
#dropping NaN values still leaves us with 140 values
print(f"Original data shape: {happy_df.shape}")
print(f"No NaN shape: {happy_df.dropna().shape}")
```
